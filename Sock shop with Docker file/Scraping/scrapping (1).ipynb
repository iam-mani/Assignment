```python
import requests
```

```python
import bs4
```

```python
res=requests.get('https://microservices-demo.github.io')
print(res.text)
```

```python
soup=bs4.BeautifulSoup(res.text,'lxml')
for link in soup.find_all('a',href=True):
  print(link['href'])
```
